## The Conjugate Gaussian Linear Model
### Basics
#### Part (D)

By "fitted line", my understanding is that this is the line of $\mathbf{y} = \mathbf{X\hat{\beta}}$, in which $\mathbf{y}$ is the dependent variable countryâ€™s GDP growth rate (GR6096), $\mathbf{X}$ is the independent variable the country's level of defense spending as a fraction of its GDP (DEF60), and $\mathbf{\hat{\beta}}$ is the mean of $\mathbf{\beta}$'s posterior distribution -- note that this mean term is the same for conditional posterior and marginal posterior, both of which doesn't have the term $\omega$. We have such mean term as $$\mathbf{\hat{\beta}} = \mathbf{U^{-1}V^T}$$ in which $\mathbf{U}=\mathbf{X^T\Lambda X+K}$, $\mathbf{V}=\mathbf{y^T\Lambda X+m^TK}$.

As the problem suggested, we set $\mathbf{\Lambda}=\mathbf{I}_{2\times 2}$, and $\mathbf{K}=\text{diag}(K_1,K_2)$ be something vague -- I pick $K_1 = K_2 = 0.1$. Moreover, I pick the prior mean of $\mathbf{\beta}$ as $\mathbf{m}=\begin{bmatrix}0.1, 0.1\end{bmatrix}^T$ -- with the precision so vague, this prior mean term won't have too much effect on posterior parameter $\mathbf{\hat{\beta}}$.

```python
gdpgrowth=pd.read_csv('gdpgrowth.csv')
n = gdpgrowth.shape[0]
y = np.array(gdpgrowth['GR6096']).reshape(n,1)
X = np.concatenate([np.ones((n,1)), np.array(gdpgrowth['DEF60']).reshape(n,1)], axis=1)
```

```python
def basic_normal_gamma(X, y, m, K):
    Lambda = np.identity(X.shape[0])
    U = (X.T).dot(Lambda).dot(X)+K
    V = (y.T).dot(Lambda).dot(X)+(m.T).dot(K)
    Beta_hat = np.linalg.inv(U).dot(V.T)
    return Beta_hat
```

```python
Beta_hat = basic_normal_gamma(X=X, y=y, m=np.array([[.1,.1]]).T, K=np.diag([.1,.1]))
```

```python
basic_normal_gamma(X=X, y=y, m=np.array([[0.,0.]]).T, K=np.diag([0.,0.]))
```

```python
# OLS fit
beta_OLS = np.linalg.inv((X.T).dot(X)).dot(X.T).dot(y)
beta_OLS
```

Plot fitted line along with scatter plot of original point, also adding in the OLS fit later -- note that the posterior mean of $\beta$ when setting all hyperparameters to $0$ or $\mathbf{0}$ is exactly the same as the OLS $\beta$:

```python
plt.scatter(X[:,1], y, s=7)
plt.plot(X[:,1], X.dot(Beta_hat), color='mediumseagreen', linewidth=.9, label='Basic NG')
plt.plot(X[:,1], X.dot(np.linalg.inv((X.T).dot(X)).dot(X.T).dot(y)), color='orange', linewidth=.9, label='OLS')
plt.legend(loc='best')
plt.title(r"Fitted Line Using Posterior $\mathbf{\beta}$ Mean with Uninformative Prior")
plt.xlabel("DEF60")
plt.ylabel("GR6096")
```

### A Heavy-Tailed Error Model
#### Part (C)

We have the posterior conditional distributions specified in notes, thus we go straight to code up the Gibbs Sampler:

```python
# # raw-code version
# # Initialize hyperparameters
# p = X.shape[1]
# m, K, d, eta, h, iteration, seed = np.array([.1,.1]).T, np.diag([.1]*X.shape[1]), 1., 1., 1., 1000, 2019
# # Sample Lambda, omega and beta from their posterior conditional distributions respectively: for Gibbs Sampling,
# # all parameters depend on each other, thus we need to start with some parameter having some value already -- I 
# # pick Lambda to start as identity matrix, for convenience purpose
# np.random.seed(seed=seed) # check if we wrote this line, do we still need to write random_state=seed in .rvs()
# omega_post, beta_post, Lambda_post = np.zeros((iteration,)), np.zeros((iteration,X.shape[1])), np.zeros((iteration,X.shape[0],X.shape[0]))
# for i in range(iteration):
#     # We sample omega and beta using the same function, since their posterior parameters share some non-trivial 
#     # terms -- we can save ourselves from writing repetitive code by doing this.
#     omega_post[i], beta_post[i] = sample_omega_and_beta(X=X, Lambda=np.identity(X.shape[0]), d=d, eta=eta, K=K, y=y, m=m) if i==0 \
#                             else sample_omega_and_beta(X=X, Lambda=Lambda_post[i-1], d=d, eta=eta, K=K, y=y, m=m)
#     Lambda_post[i] = np.diag([*map(lambda j: sample_Lambda_j(X_j=X[j,:], y_j=y[j,0], \
#                 h=h, omega_post=omega_post[i], beta_post=beta_post[i]), np.arange(X.shape[0]))])
```

```python
def sample_omega_and_beta(X, Lambda, d, eta, K, y, m):
    U = (X.T).dot(Lambda).dot(X)+K
    V = (y.T).dot(Lambda).dot(X)+(m.T).dot(K)
    post_omega_a = (X.shape[0]+d)/2.
    q = (y.T).dot(Lambda).dot(y)+(m.T).dot(K).dot(m)-V.dot(np.linalg.inv(U)).dot(V.T)
    post_omega_b = (eta+q)/2.
    omega_post = sps.gamma.rvs(a=post_omega_a, scale=1./post_omega_b, size=1)[0]
    post_beta_mean = np.linalg.inv(U).dot(V.T)
    post_beta_cov = np.linalg.inv(omega_post*U)
    beta_post = sps.multivariate_normal.rvs(mean=post_beta_mean.flatten(), cov=post_beta_cov, size=1)
    # note that beta_post is a (p,)-shaped 1d array -- need to be cautious when calculating Lambda_i's parameter
    return omega_post, beta_post

def sample_Lambda_j(X_j, y_j, h, omega_post, beta_post):
    post_Lambdaj_a = (h+1)/2.
    post_Lambdaj_b = (omega_post*np.power(y_j-X_j.dot(beta_post.T),2)+h)/2.
    Lambdaj_post = sps.gamma.rvs(a=post_Lambdaj_a, scale=1./post_Lambdaj_b, size=1)[0]
    return Lambdaj_post

def gibbs_sampler_heavy_tailed(X, y, m, K, d, eta, h, iteration, burn_in, thinning, seed):
    # Sample Lambda, omega and beta from their posterior conditional distributions respectively: for Gibbs Sampling,
    # all parameters depend on each other, thus we need to start with some parameter having some value already -- I 
    # pick Lambda to start as identity matrix, for convenience purpose
    np.random.seed(seed=seed) # check if we wrote this line, do we still need to write random_state=seed in .rvs()
    omega_post, beta_post, Lambda_post = np.zeros((iteration,)), np.zeros((iteration,X.shape[1])), np.zeros((iteration,X.shape[0],X.shape[0]))
    for i in range(iteration):
        # We sample omega and beta using the same function, since their posterior parameters share some non-trivial 
        # terms -- we can save ourselves from writing repetitive code by doing this.
        omega_post[i], beta_post[i] = sample_omega_and_beta(X=X, Lambda=np.identity(X.shape[0]), d=d, eta=eta, K=K, y=y, m=m) if i==0 \
                            else sample_omega_and_beta(X=X, Lambda=Lambda_post[i-1], d=d, eta=eta, K=K, y=y, m=m)
        Lambda_post[i] = np.diag([*map(lambda j: sample_Lambda_j(X_j=X[j,:], y_j=y[j,0], \
                    h=h, omega_post=omega_post[i], beta_post=beta_post[i]), np.arange(X.shape[0]))])
    return omega_post[burn_in:][0::thinning], beta_post[burn_in:][0::thinning], Lambda_post[burn_in:][0::thinning]

def make_plot_posterior_fitted_line(X, y, heavy_tailed_gibbs_beta, basic_beta):
    plt.scatter(X[:,1], y, s=7)
    plt.plot(X[:,1], X.dot(heavy_tailed_gibbs_beta), color='r', linewidth=.9, label='Heavy-Tailed NG')
    plt.plot(X[:,1], X.dot(basic_beta), color='mediumseagreen', linewidth=.9, label='Basic NG')
    plt.plot(X[:,1], X.dot(np.linalg.inv((X.T).dot(X)).dot(X.T).dot(y)), color='orange', linewidth=.9, label='OLS')
    plt.legend(loc='best')
    plt.title(r"Fitted Lines Using Posterior $\mathbf{\beta}$ Mean with Uninformative Prior")
    plt.xlabel("DEF60")
    plt.ylabel("GR6096")
```

I got drastically different results from setting the hyperparameters differently:

```python
hyperparam_m, hyperparam_K = np.array([[.1, .1]]), np.diag([.1, .1])
omega_post, beta_post, Lambda_post = gibbs_sampler_heavy_tailed(X=X, y=y, \
                                        m=hyperparam_m.T, K=hyperparam_K, d=4., eta=4., h=4., \
                                        burn_in=2000, thinning=3, iteration=5000, seed=2019)
print(np.mean(omega_post), np.mean(beta_post,axis=0))
Beta_hat = basic_normal_gamma(X=X, y=y, m=hyperparam_m.T, K=hyperparam_K)
make_plot_posterior_fitted_line(X=X, y=y, heavy_tailed_gibbs_beta=np.mean(beta_post,axis=0), basic_beta=Beta_hat)
```

##### Initial Observation:
I initialized my hyperparameters as: $\mathbf{m}=\begin{bmatrix}0.1,0.1\end{bmatrix}^T$, $\mathbf{K}=\text{diag}(0.1,0.1)$, $d=1$, $\eta=1$, $h=1$. However, no matter after how much iteration, the fitted line varies a lot after each iteration -- my plots of iteration$=5000$, $5001$, $5002$ look sharply different from each other. Before checking my implementation, I started to suspect if it was because of my initialization, as this was what happened last time I implemented Gibbs Sampler on Abhra's class. I gradually modified $\mathbf{K}, h, d$, in which the first two didn't change much, change in $d$ seems to decrease the differences after each iteration but still didn't quite solve the problem. Finally, I modified $\mathbf{m}$ to current value, and with only $2000$ iterations I had three plot looks almost identical except the intercept location, and all look robust against the outliers. 

My current hyperparameter setting:

$\mathbf{m}=\begin{bmatrix}1,1\end{bmatrix}^T$, $\mathbf{K}=\text{diag}(2,2)$, $d=5$, $\eta=1$, $h=4$.

##### Update on Above Comments:

For Gibbs Sampling, we are not just looking at the very last sample -- instead, we need to take the mean of all (burnt-in and thinned) samples to derive our posterior estimate of parameters. After we did that, the sample mean of posterior samples doesn't change much.
