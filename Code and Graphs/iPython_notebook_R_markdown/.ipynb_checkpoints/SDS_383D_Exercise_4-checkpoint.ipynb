We have the posterior distribution of each $\theta_i$ as $$\theta_i|y_{i,1:n_i}\sim\mathcal{N}\Big(\frac{\sum y_{ij}+\omega\cdot\mu}{n_i+\omega}, \lambda^{-1}(n_i+\omega)^{-1}\Big)$$ in which $\lambda=1/\sigma^2$, $\omega=1/\tau^2$; also $\sum y_{ij}$ and $n_i$ are sum of total math score and count of students from School $i$, respectively, we can directly calculate these two from the Pandas dataframe directly, as well as the mean $\bar{y}_i$, which we could directly use later conveniently.

```python
# note that Panda's var function will compute the sample variance, thus is the sum of squares divided by (n-1)
mathtest_by_school_df = mathtest_df.groupby('school').agg(['sum', 'count', 'mean', 'var'])['mathscore']
mathtest_by_school_df.columns = ['sum_by_school', 'student_count_by_school', 'mean_by_school', 'variance_by_school']
mathtest_by_school_df['variance_by_school'] = mathtest_by_school_df['variance_by_school'] \
        *(mathtest_by_school_df['student_count_by_school']-1)/mathtest_by_school_df['student_count_by_school']
mathtest_by_school_df.head()
```

##### Empirical Bayes Method

One way to fit the normal hierarchical model is by using the conditionals given by the question to derive the joint distribution, and integrating out $\theta^i$'s to get the marginal distribution of $y_{ij}$'s given hyperparameters $\mu, \sigma, \tau$, and using it as an objective function to optimize w.r.t. the hyperparameters. I implemented Nelder-Mead method:

```python
# Compute Posterior Mean and Variance
def compute_posterior_mean_and_variance(mu, lamda, omega, mathtest_by_school):
    """Compute posterior mean and variance for each theta_i.
    
    Args:
        mu: one of the hyperparameters.
        omega: one of the hyperparameters.
        lambda: one of the hyperparameters.
        mathtest_by_school: a 2d numpy array with total math test score, count of students, and mean score, for each school.
    
    Return:
        two numpy arrays -- posterior mean and variance for each theta_i.
    """
    posterior_mean = (mathtest_by_school[:,0]+omega*mu)/(mathtest_by_school[:,1]+omega)
    posterior_var = 1./(lamda*(mathtest_by_school[:,1]+omega))
    return posterior_mean, posterior_var
```

```python
empirical_lambda, empirical_omega = 1./empirical_sigma2, 1./empirical_tau2
empirical_mu = np.mean(mathtest_by_school[:,2])
# compute posterior mean and variance
mu, omega, lamda = empirical_mu, empirical_omega, empirical_lambda
posterior_mean, posterior_var = compute_posterior_mean_and_variance(mu, omega, lamda, mathtest_by_school)
```

##### Try Nelder-Mead Method to Optimize Hyperparameters $\mu$, $\lambda$, $\omega$ Posterior Log-Likelihood

The posterior likelihood is computed by taking the sum of all log-likelihood of each $y_{ij}$ with mean being the posterior mean of corresponding $\theta_i$, and variance being the hyperparameter $1/\lambda$.

```python
def Nelder_Mead_for_mathtest(x):
    """Write specific function for Nelder-Mead optimization over combination of hyperparameters [mu, lamda, omega] -- the optimization 
       is meant to minimize the given function, but our goal is to maximize the marginal likelihood, thus we put a negative sign 
       in front to achieve the optimization direction.
       
       Args:
           x: a list/array contains [b1, b2, tau_1_sqd]
           
       Returns:
           negative marginal likelihood with constants removed
    
    """
    
    mu, lamda, omega = x[0], x[1], x[2]
    posterior_mean = (mathtest_by_school[:,0]+omega*mu)/(mathtest_by_school[:,1]+omega)
    posterior_var = 1./(lamda*(mathtest_by_school[:,1]+omega))
    total_marginal_logpdf = np.sum([*map(lambda i: (mathtest_by_school[i,1]/2.)*np.log(lamda)+.5*np.log(omega/(mathtest_by_school[i,1]+omega)) \
                    -.5*(lamda*mathtest_squaredsum_by_school[i]+lamda*omega*mu*mu-lamda*np.power(mathtest_by_school[i, 0]+omega*mu, 2) \
                    /(mathtest_by_school[i,1]+omega)), np.arange(mathtest_by_school.shape[0]))])
    return -total_marginal_logpdf
```

##### Gibbs Sampling

Another way to do it is what we've heavily studied from our MCMC class this semester, the Gibbs sampler: assume appropriate prior distributions for $\mu$, $\lambda=1/\sigma^2$ and $\omega=1/\tau^2$, derive the posterior conditionals for each parameter, and iteratively sample each parameter over a big enough amount of iterations, then take the mean of samples of each parameter over all iterations as our estimate of such parameter. We implement the Markov chain as follows:

```python
def sample_theta_i(mathtest_by_school, mu, lamda, omega, n):
    posterior_mean = (mathtest_by_school[:,0]+omega*mu)/(mathtest_by_school[:,1]+omega)
    posterior_var = 1./(lamda*(mathtest_by_school[:,1]+omega))
    return np.array([*map(lambda i: sps.norm.rvs(loc=posterior_mean[i], scale=np.sqrt(posterior_var[i])), np.arange(n))])

def sample_mu(theta_i, mu, lamda, omega, nu, phi, n):
    precision=omega*lamda*n+phi
    return sps.norm.rvs(loc=(omega*lamda*np.sum(theta_i)+phi*nu)/precision, scale=np.sqrt(1./precision))

def sample_lambda(mathtest_df, mathtest_by_school, theta_i, mu, omega, a, b, total_m, n):
    double_sum = np.sum([*map(lambda i: np.sum(np.power(np.array(mathtest_df[mathtest_df['school']==i+1]['mathscore'])-theta_i[i], 2)), np.arange(n))])
    single_sum = np.sum(np.power(theta_i-mu, 2))
    rate = (double_sum+omega*single_sum)/2.+b
    return sps.gamma.rvs(a=(total_m+n)/2.+a, scale=1./rate)

def sample_omega(theta_i, mu, lamda, c, d, n):
    another_single_sum = np.sum(np.power(theta_i-mu, 2))
    another_rate = lamda*another_single_sum/2.+d
    return sps.gamma.rvs(a=n/2.+c, scale=1./another_rate)

# a, b, c, d, nu, phi = 1., 1., 1., 1., np.mean(mathtest_df['mathscore']), 10.
# seed = 2019
# burn_in, tot_iter = 1000, 2000
def Markov_chain_Gibbs_sampler_mathtest(mathtest_df, mathtest_by_school, a, b, c, d, nu, phi, burn_in, tot_iter, seed):
    np.random.seed(seed)
    # prepare Markov chain by initialize arrays to store burn_in and actual samples of all parameters
    total_m = mathtest_df.shape[0] # number of total students
    n = mathtest_by_school.shape[0] # number of schools
    empirical_sigma2 = np.mean(mathtest_by_school[:,3])
    empirical_tau2 = np.var(mathtest_by_school[:,2])/empirical_sigma2
    empirical_lambda, empirical_omega = 1./empirical_sigma2, 1./empirical_tau2
    empirical_mu = np.mean(mathtest_by_school[:,2])
    theta_i_burn_in, mu_burn_in, lambda_burn_in, omega_burn_in = np.zeros(n), empirical_mu, empirical_lambda, empirical_omega
    theta_i_array, mu_array, lambda_array, omega_array = np.zeros((tot_iter-burn_in, n)), np.zeros(tot_iter-burn_in), \
                                                            np.zeros(tot_iter-burn_in), np.zeros(tot_iter-burn_in)
    for idx in range(burn_in):
        theta_i_burn_in = sample_theta_i(mathtest_by_school, mu_burn_in, lambda_burn_in, omega_burn_in, n)
        mu_burn_in = sample_mu(theta_i_burn_in, mu_burn_in, lambda_burn_in, omega_burn_in, nu, phi, n)
        lambda_burn_in = sample_lambda(mathtest_df, mathtest_by_school, theta_i_burn_in, mu_burn_in, omega_burn_in, a, b, total_m, n)
        omega_burn_in = sample_omega(theta_i_burn_in, mu_burn_in, lambda_burn_in, c, d, n)

    for idx in range(tot_iter-burn_in):
        theta_i_array[idx] = sample_theta_i(mathtest_by_school, mu_burn_in, lambda_burn_in, omega_burn_in, n) if idx==0 \
                            else sample_theta_i(mathtest_by_school, mu_array[idx-1], lambda_array[idx-1], omega_array[idx-1], n)
        mu_array[idx] = sample_mu(theta_i_burn_in, mu_burn_in, lambda_burn_in, omega_burn_in, nu, phi, n) if idx==0 \
                        else sample_mu(theta_i_array[idx], mu_array[idx-1], lambda_array[idx-1], omega_array[idx-1], nu, phi, n)
        lambda_array[idx] = sample_lambda(mathtest_df, mathtest_by_school, theta_i_burn_in, mu_burn_in, omega_burn_in, a, b, total_m, n) if idx==0 \
                            else sample_lambda(mathtest_df, mathtest_by_school, theta_i_array[idx], mu_array[idx], omega_array[idx-1], a, b, total_m, n)
        omega_array[idx] = sample_omega(theta_i_burn_in, mu_burn_in, lambda_burn_in, c, d, n) if idx==0 \
                            else sample_omega(theta_i_array[idx], mu_array[idx], lambda_array[idx], c, d, n)
        
    return theta_i_array, mu_array, lambda_array, omega_array
```

```python
t = time.time()
# a, b, c, d, nu, phi = 1., 1., 1., 1., np.mean(mathtest_df['mathscore']), 10.
a, b, c, d, nu, phi = 1e-3, 1e-3, 1e-3, 1e-3, np.mean(mathtest_df['mathscore']), .1
burn_in, tot_iter, seed = 2000, 4000, 2019
theta_i_array, mu_array, lambda_array, omega_array = Markov_chain_Gibbs_sampler_mathtest(mathtest_df, mathtest_by_school, \
                                                            a, b, c, d, nu, phi, burn_in, tot_iter, seed)
elapsed = time.time() - t
print("The chain ran for %.2f seconds." % elapsed)
```
